```python
import pandas as pd
import numpy as np



# Define lists to hold the scraped data
activity_types = []
descriptions = []
locations = []
show_times = []
prices = []

# Locate all the main sections (h3 tags) and their subsequent details
for section in soup.find_all('h3'):
    # Get the activity type from the h3 text
    activity_type = section.get_text(strip=True)
    
    # Initialize details for this section
    about = np.nan
    location = np.nan
    show_time = np.nan
    price = np.nan
    
    # Find the next siblings, where the details are located
    sibling = section.find_next_sibling()
    
    while sibling and sibling.name not in ['h3', 'h2']:
        # Check for different types of tags for various details
        if sibling.name == 'ul':
            for li in sibling.find_all('li'):
                if 'About:' in li.get_text():
                    about = li.get_text(strip=True).replace('About: ', '')
                elif 'Location:' in li.get_text():
                    location = li.get_text(strip=True).replace('Location: ', '')
                elif 'Show-times:' in li.get_text() or 'Opening times:' in li.get_text():
                    show_time = li.get_text(strip=True).replace('Show-times: ', '').replace('Opening times: ', '')
                elif 'Pricing:' in li.get_text():
                    price = li.get_text(strip=True).replace('Pricing: ', '')
        
        # Move to the next sibling element
        sibling = sibling.find_next_sibling()
    
    # Append the scraped details to the lists
    activity_types.append(activity_type)
    descriptions.append(about)
    locations.append(location)
    show_times.append(show_time)
    prices.append(price)

# Create a DataFrame to hold the scraped data
df = pd.DataFrame({
    'Activity_Type': activity_types,
    'Description': descriptions,
    'Location': locations,
    'Show_Time': show_times,
    'Price': prices
})

# Replace any empty string with NaN for consistency
df.replace('', np.nan, inplace=True)

# Save the DataFrame to a CSV file
df.to_csv('cairo_activities.csv', index=False)

print("Data has been scraped and saved to 'cairo_activities.csv'")
```
